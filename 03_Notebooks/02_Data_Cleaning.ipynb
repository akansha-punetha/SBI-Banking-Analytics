# Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df_accounts = pd.read_csv("../01_Raw_Data/accounts.csv")
df_branches = pd.read_csv("../01_Raw_Data/branches.csv")
df_cards = pd.read_csv("../01_Raw_Data/cards.csv")
df_customers = pd.read_csv("../01_Raw_Data/customers.csv")
df_loans = pd.read_csv("../01_Raw_Data/loans.csv")
df_merchants = pd.read_csv("../01_Raw_Data/merchants.csv")
df_transactions = pd.read_csv("../01_Raw_Data/transactions.csv")
```

```python
df_customers["created_at"] = pd.to_datetime(df_customers["created_at"])
df_accounts["open_date"] = pd.to_datetime(df_accounts["open_date"])
df_transactions["transaction_date"] = pd.to_datetime(df_transactions["transaction_date"])
df_cards["expiration_date"] = pd.to_datetime(df_cards["expiration_date"])
df_loans["start_date"] = pd.to_datetime(df_loans["start_date"])
```

```python
df_customers.to_csv("../02_Cleaned_Data/customers.csv", index=False)
df_accounts.to_csv("../02_Cleaned_Data/accounts.csv", index=False)
df_transactions.to_csv("../02_Cleaned_Data/transactions.csv", index=False)
df_cards.to_csv("../02_Cleaned_Data/cards.csv", index=False)
df_loans.to_csv("../02_Cleaned_Data/loans.csv", index=False)
df_branches.to_csv("../02_Cleaned_Data/branches.csv", index=False)
df_merchants.to_csv("../02_Cleaned_Data/merchants.csv", index=False)

print("All cleaned datasets exported successfully.")
```
